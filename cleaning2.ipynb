```python
import pandas as pd 
```

```python
music = pd.read_csv('/home/jovyan/MonDossier/data/my_file (1).csv',index_col = 0)
music
```

```python
music["Peak"]=music["Peak"].astype(str)
```

```python
#regex = true 
```

```python
music["Peak"]=music["Peak"].replace('\[', '', regex=True)
music["Peak"]=music["Peak"].replace('\]', '', regex=True)
```

```python
music["All Time Peak"]=music["All Time Peak"].replace('\[', '', regex=True)
music["All Time Peak"]=music["All Time Peak"].replace('\]', '', regex=True)
```

```python
s = "The 1989 World Tour   Sticky & Sweet Tour ‡[4][a] "
import re 
pattern =  r"([A-Za-z0-9\s]+\&*[A-Za-z0-9\s]+)"
match = re.search(pattern,s)

    
print(match.group(0))
print(match.group(1))

```

```python
music["Tour title"] = music["Tour title"].str.extract(r'([A-Za-z0-9\s]+\&*[A-Za-z0-9\s]+)', expand=False)
```

```python
music["Ref."]=music["Ref."].replace('\[', '', regex=True)
music["Ref."]=music["Ref."].replace('\]', '', regex=True)
```

```python
music.columns=['Peak', 'All Time Peak', "Actual gross",
       'Adjusted gross (in 2022 dollars)', 'Artist', 'Tour title', 'Year(s)',
       'Shows', 'Average gross', 'Ref.']
```

```python
music["Actual gross"]=music["Actual gross"].replace('\[', '', regex=True)
music["Actual gross"]=music["Actual gross"].replace('\]', '', regex=True)
```

```python
l = "$229,100,000b "
import re 
pattern =  r"(\$\w+.*\d)"
match = re.search(pattern,l)

    
print(match.group(0))
```

```python
music["Actual gross"]= music["Actual gross"].str.extract(r"(\$\w+.*\d)", expand=False)
```

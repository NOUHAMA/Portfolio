```python
#read file and delete front rows (3 rows) 
import pandas as pd 

cola = pd.read_excel('/home/jovyan/MonDossier/Copie de Cola.xls', skiprows=3)
cola
```

```python
cola.columns
```

```python
#rename first column : financial statement
new_column_names = ['financial statement', "Metrics", "2009","2010", "2011","2012","2013","2014","2015","2016","2017","2018"]
cola.columns = new_column_names
cola
```

null values 

```python
#are all values in the first column are null : false 
null_count_C =cola["financial statement"].isnull().all()
print(null_count_C)
```

```python
#insert values in the first column (repeat each value * times)
data = ["Profit & Loss statement"] * 18 + ['Balance Sheet'] * 44 + ["Cash Flow statement"] * 38
```

```python
cola["financial statement"] = data
cola
```

```python
#check other null values 
cola.isnull()
```

```python
#define threshold
#delete rows that have at least 11 rows null
cola.dropna(thresh=11, inplace=True)
cola
```

format and data type

```python
print(cola.dtypes)
```

```python
cola.columns
```

```python
#convert columns with numeric data 
columns_to_convert = ['2009', '2010', '2011', '2012',
       '2013', '2014', '2015', '2016', '2017', '2018']


for col in columns_to_convert:
    cola[col] = pd.to_numeric(cola[col], errors='coerce')
```

```python
print(cola.dtypes)
```

```python
cola.isnull().any()
```

```python
#fill the null values with 0 
cola.fillna(0) 
```

```python
#duplicates 
cola.duplicated().sum()
```

```python
#change the structure of the data frame 
formatted_cola = pd.melt(cola, id_vars=["financial statement","Metrics"], var_name="year", value_name="amount")
formatted_cola
```

```python
#verify that all metrics 
null_values = formatted_cola["amount"].isnull().any()
print(null_values)
```

```python
mask = formatted_cola["Metrics"] == "in million USD"
df = formatted_cola.loc[mask, :]
df
```

```python
#The ~ operator negates the mask, meaning it selects rows that do not meet the condition
cola_cleaned = formatted_cola[~mask]
cola_cleaned
```

```python
cola_cleaned.to_excel('cola_clean_version.xlsx', index=False)
#index = false to prevent pandas from writting row indices 
print("yes")
```
